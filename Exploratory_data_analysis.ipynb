```python
arr_txt["Age"].fillna(arr_txt["Age"].median(),inplace=True)
arr_txt["Embarked"].fillna(arr_txt["Embarked"].mode(),inplace=True)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="PassengerId",y="Gender",data=arr_txt,palette="pastel")
plt.title("SCATTER PLOT",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(6,7))
sns.barplot(x="Ticket",y="Fare",data=arr_txt,palette="coolwarm")
plt.title("BARPLOT",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.boxplot(x="Age",y="Fare",data=arr_txt,palette="colorblind")
plt.title("BOXPLOT",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.violinplot(x="Cabin",y="Fare",data=arr_txt,palette="dark")
plt.title("VIOLIN PLOT",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(3,2))
sns.lineplot(x="PassengerId",y="Ticket",data=arr_txt,palette="pastel")
plt.title("LINE PLOT",fontsize=8)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.histplot(x="Fare",y="Gender",data=arr_txt,palette="pastel")
plt.title("HISTOGRAM PLOT",fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.kdeplot(x="PassengerId",y="Pclass",data=arr_txt,palette="pastel")
plt.title("DENSITY PLOT",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Fare",data=arr_txt,palette="pastel")
plt.title("COUNT PLOT",fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.pie(x="Age",data=arr_txt)
plt.title("PIE PLOT",fontsize=14)
plt.show()
```
